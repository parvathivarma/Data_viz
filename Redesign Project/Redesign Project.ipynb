## Description 


The chart can be filtered on the basis of skill's current demand in job market. A skill with high demand are those skillsets which has an increasing demand over the year and is among the top 10 skill sets. The skill with decreasing demand are those skills which are losing its position in the market due to multiple reasons. One such reason is that the people with the skillset is more than the job vacancies. The third group is the one with skills that are no longer required in the market according to LinkedIn's Top 25 Skills list. The number beneath the skill name indicates the difference in rank for a skill compared to previous year.


## Claim


Increasing demand in data skills have created even more demand for other data driven fields
